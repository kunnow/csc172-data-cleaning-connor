```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw_dataset.csv")
```

```python
print("Explolatory Checks")
print("\nData Info:")
df.info()

print("\nSummary Statistics (Before Cleaning):")
display(df.describe())

print("\nMissing Values per Column:")
print(df.isnull().sum())

print("\nTotal Duplicates:", df.duplicated().sum())
```

```python
shape_before = df.shape
sample_before = df.head()
summary_before = df.describe()
```

```python
print("\nCleaning Steps")
```

```python
for col in df.columns:
    if df[col].dtype in ["float64", "int64"]:
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
for col in df.select_dtypes(include=[np.number]).columns:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
    df[col] = np.clip(df[col], lower, upper)
```

```python
print("\nSnapshots Comparison")
print("Shape Before:", shape_before)
print("Shape After:", df.shape)

print("\nSample Rows Before Cleaning:")
display(sample_before)

print("\nSample Rows After Cleaning:")
display(df.head())

print("\nSummary Statistics (After Cleaning):")
display(df.describe())
```

```python
df.to_csv("../data/cleaned_dataset.csv", index=False)
print("\nCleaned dataset saved as: data/cleaned_dataset.csv")
```
